# Visualisation of the Random Forest model

```python
# the usual imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, Normalizer
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
```

```python
# load the data
df = pd.read_csv("newdata.csv")
```

```python
# x/y split
X = df.drop('AcademicPerformance', axis=1)
y = df['AcademicPerformance']
```

```python
from sklearn.tree import DecisionTreeClassifier 
clf = DecisionTreeClassifier()
model = clf.fit(X, y)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

# Assuming you have already trained your decision tree classifier and named it clf

# Plot decision tree
plt.figure(figsize=(24, 16))
_ = tree.plot_tree(clf, 
                   feature_names=X.columns,  
                   class_names=["Average", "Below average", "Excellent", "Good"],  # Provide class names here
                   filled=True, rounded=True, fontsize=12, node_ids=True)
```

```python
# pip install dtreeviz
import dtreeviz
viz_model = dtreeviz.model(clf,
                           X_train=X, y_train=y,
                           feature_names=X.columns,
                           target_name='AcademicPerformance',
                           class_names=["Average", "Below average", "Excellent", "Good"])

# for viewing in the Jupyter notebook
# Note: this is not practical with large 
# decision trees!
# adjust the scale if you need a bigger picture (for example 2.5)
viz_model.view(scale=1.5)
```
