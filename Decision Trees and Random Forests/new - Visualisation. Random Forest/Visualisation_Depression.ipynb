# Visualisation of the Random Forest model

```python
# the usual imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, Normalizer
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
```

```python
# load the data
df = pd.read_csv("newdata.csv")
```

```python
# x/y split
X = df.drop('DepressionStatus', axis=1)
y = df['DepressionStatus']
```

```python
from sklearn.tree import DecisionTreeClassifier 
clf = DecisionTreeClassifier()
model = clf.fit(X, y)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

# Assuming you have already trained your decision tree classifier and named it clf

# Plot decision tree
plt.figure(figsize=(24, 16))
_ = tree.plot_tree(clf, 
                   feature_names=X.columns,  
                   class_names=["No", "Sometimes", "Yes"],  # Provide class names here
                   filled=True, rounded=True, fontsize=12, node_ids=True)
```

```python
# pip install dtreeviz
import dtreeviz
viz_model = dtreeviz.model(clf,
                           X_train=X, y_train=y,
                           feature_names=X.columns,
                           target_name='DepressionStatus',
                           class_names=["No", "Sometimes", "Yes"])

# for viewing in the Jupyter notebook
# Note: this is not practical with large 
# decision trees!
# adjust the scale if you need a bigger picture (for example 2.5)
viz_model.view(scale=1.5)
```

Age plays a significant role in academic performance. We can see that younger people perform well in studies, but between 21.5 and 25 years, students show "Excellent Academic performance." The reason for this might be a more conscious attitude towards studies.

The next important node is "Number Of Friends," which quantifies the social aspect by indicating the number of friends each individual has, contributing to the understanding of social dynamics within the dataset. The node describes that all students who are good in communication and socialization have good academic performance. I think communication and negotiation skills are important in studies and can actually affect a lot.

The next one is "Face Challenges To Complete Academic Task," which explores whether individuals encounter challenges in completing academic tasks. There are people who do not have any challenges and who see studies as one big challenge :) The former ones perform better in studies.

"Sleep Per Day Hours" is the next node that provides insights that sleep does not have a significant impact on academic performance because all students are about to have "good" but possess different amounts of sleeping hours.

The last node is "Depression Status," which shows that people with no depression symptoms perform better at school. I think it is self-explanatory.
